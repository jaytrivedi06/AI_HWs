```python
import numpy as np
import random
import copy
import collections
```

```python
def numInput(number):                                                          #### to check whether a user input is a valid number or not
    if number.isdigit() and int(number) > 0:
        return int(number)
    else:
        print ("You must enter a postive integer:")
        a = input()
        return numInput(a)
        
```

```python
def env(shape = 1, k = 1):                                                            #### create an environment of N dimension
    if k!=1:
        shape = numInput(input("Enter dimensions of the environment : "))
        dim = []
        for i in range(1,shape+1):
            dim.append(numInput(input("Enter dimension "+str(i)+" : ")))

    else:
        dim = [2]        
    return np.zeros(tuple(dim), dtype = int)
```

```python
def dirt(y, k = 1):                                                  #### get dirt coordinates
    if k != 1:  
        n = numInput(input("Enter number of dirty squares : "))
        while n > y.size:
            print("Please enter a number less than or equal to",y.size)
            n = numInput(input("Enter number of dirty squares : "))                                                             
        print("Please enter indexes in comma seperated format. For exampel 1d environment write 1, for 2d write 1,2")
        for i in range(n):
            index = tuple(map(int,input("Enter Indices of dirty square "+str(i+1)+": ").split(',')))
            if len(index) == y.ndim:
                np.put(y , index, 1)
            else:
                print("Number of indices not equal to dimension of environment thus input is discarded")

    else:
        index = random.choice([[0],[1],[0,1],[]])
        for i in index:
            np.put(y , index, 1)
    return y
```

```python
def performance(environment ):
    ### The performance measure awards one point for each clean square at each time step,
    ### over a “lifetime” of 1000 time steps and averages it
    return int(environment.size - np.sum(environment))
```

```python
def agent(y):
    loc = []
    for i in list(y.shape):
        loc.append(random.randint(0,i-1))
    return loc 

def percept(y , loc):
    return y[loc]
```

```python
def action (y,loc ,percept):
    if percept == 1:
        np.put(y,loc,0)
        output = "Sucked Dirt"
    elif percept == 0 and loc == 0:
        loc = 1
        output = "Moved Right"
    else :
        loc = 0
        output = "Moved Left"
    return y ,loc ,output
```

```python
for i in range(25):
    f = env()
    y = dirt(f)
    l = agent(y)
    print ("Agent Initial Location ", l)
    print ("Initial dirt configuration ", y)
    o_seq = []
    per = 0
    for t in range(1000):
        p = percept(y,l)
        y, l, o = action(y,l,p)
        o_seq.append(o)
        per = per + performance(y)
    
    print("Performance ",per)
    print ("Average Performance ", per/1000)
    print("Output sequence containing first 10 steps ",o_seq[0:10])
    print(" ==================================== ")
```
